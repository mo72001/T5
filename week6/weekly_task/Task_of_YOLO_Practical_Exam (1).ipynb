## Process Video Frames and Identify Vehicles Left the Road Form Each Street

For each frame in the video, use the YOLO model to detect and track vehicles. You will need to write a loop that processes each frame and applies the YOLO model.

In each frame, after detecting the vehicles, determine whether they exited the road via the top, bottom, left, or right street. You can use the positions of the bounding boxes provided by YOLO to do this.

```python
# this cide from the lab of Object_Counting_YOLO in section Define the Line in the Middle of the Frame
mid_x = width // 2

mid_y = height // 2

line_points1 = [(mid_x, 20), (mid_x, height - 20)]

line_points2 = [(20, mid_y), (width - 20, mid_y)]
```

```python
# source: This code from the lab of Object_Counting_YOLO in section Initialize the Object Counter
counter1 = solutions.ObjectCounter(
    view_img=True,
    reg_pts=line_points1,
    names=model.names,
    draw_tracks=True,
    line_thickness=1,)
```

```python
# source: This code from the lab of Object_Counting_YOLO in section Initialize the Object Counter
counter2 = solutions.ObjectCounter(
    view_img=True,
    reg_pts=line_points2,
    names=model.names,
    draw_tracks=True,
    line_thickness=1,)
```

```python
# the sources if this code "I Edit it"
# https://pythonwife.com/object-tracking-in-opencv/
# https://www.geeksforgeeks.org/python-opencv-cv2-puttext-method/
# https://roboflow.com/use-opencv/draw-text-with-cv2-puttext
# https://www.geeksforgeeks.org/python-opencv-cv2-rectangle-method/
# https://docs.ultralytics.com/guides/object-counting/#real-world-applications
# https://thepythoncode.com/article/real-time-vehicle-tracking-and-counting-with-yolov8-opencv

# also I take a bet of labs

# from the lab of Object_Counting_YOLO in section Process the Video Frames
# from the lab of Motion_Detection in section Processing the Video Frame by Frame
```

```python
while cap.isOpened():
    success, image = cap.read()
    if not success:
        print('Video frame is empty or video processing has been successfully completed.')
        break

    text = 'Mohammed alrowais'

    org = (500, 40)

    fontScale = 1

    color = (255, 255, 255)

    thickness = 3

    cv2.putText(image, text, org, cv2.FONT_HERSHEY_SIMPLEX, fontScale, color, thickness)

    color2 = (0, 255, 0)

    thickness = 2

    cv2.rectangle(image, roi_top[:2], roi_top[2:], color2, thickness)

    cv2.rectangle(image, roi_bottom[:2], roi_bottom[2:], color2, thickness)

    cv2.rectangle(image, roi_left[:2], roi_left[2:], color2, thickness)

    cv2.rectangle(image, roi_right[:2], roi_right[2:], color2, thickness)

    tracks = model.track(image, persist=True, show=False)

    image = counter1.start_counting(image, tracks)

    image = counter2.start_counting(image, tracks)


    output.write(image)
```

```python
# source: This code from the lab of Object_Counting_YOLO in section Release Resources
cap.release()
output.release()
cv2.destroyAllWindows()
```
